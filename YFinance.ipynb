```python
import yfinance
import requests_html
import yahoo_fin.stock_info as si
tickers = si.tickers_dow()
num = 0
for x in tickers:
    num += 1

print(num)
print(tickers)
```

```python
all_stock_data = {}

combined_data = []
for ticker in tickers:
    try:
        data = yfinance.download(ticker, start='2015-01-01', end='2025-01-01')
        data['Ticker'] = ticker
        all_stock_data[ticker] = data
    except Exception as e:
        print(f"Failed to fetch data for {ticker}: {e}")


for ticker, data in all_stock_data.items():
    data.to_csv(f"{ticker}_stock_data.csv")
    break
        

print(all_stock_data)
```

```python
with open('all_stock_data_combinednumba2.csv', 'w') as f:
    # Loop through the stock data for each ticker
    first_ticker = True  # A flag to handle header writing
    for ticker, data in all_stock_data.items():
        # If it's the first ticker, write the header (column names)
        if first_ticker:
            data.to_csv(f, header=True)  # Write header
            first_ticker = False
        else:
            # For subsequent tickers, append without writing the header
            data.to_csv(f, header=False)

print("All stock data has been saved to 'all_stock_data_combined.csv'.")
```
